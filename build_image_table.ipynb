```python
import pandas as pd
import numpy as np

from os.path import exists
from IPython.core.display import display, HTML

pd.set_option('display.max_colwidth', None)
```

```python
def get_figs(df):
    tess_figs = []
    gaia_figs = []
    
    for tic in df.index:
        tess_path = 'tess_figs/TIC ' + str(tic) + '.png'
        gaia_path = 'gaia_figs/TIC ' + str(tic) + '.png'
        
        if exists(tess_path):
            tess_figs.append(tess_path)
        else:
            tess_figs.append(np.nan)

        if exists(gaia_path):
            gaia_figs.append(gaia_path)
        else:
            gaia_figs.append(np.nan)
            
    return tess_figs, gaia_figs
```

```python
def path_to_html(path):
    w = 300
    return '<img src="'+ path + '" width="' + str(w) + '" >'
```

```python
df = pd.read_csv('tess_gaia.csv').set_index('TIC_ID')
```

```python
tess_figs, gaia_figs = get_figs(df)
df['TESS_FIG'] = tess_figs
df['GAIA_FIG'] = gaia_figs
```

```python
fig_cols = ['TESS_FIG', 'GAIA_FIG']

format_dict = {}
for fig_col in fig_cols:
    format_dict[fig_col] = path_to_html


html = df.to_html(escape=False, formatters=format_dict)
```

```python
display(HTML(html))
```
